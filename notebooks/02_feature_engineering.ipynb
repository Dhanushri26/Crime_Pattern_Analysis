```python
import pandas as pd

# Load processed data
df = pd.read_csv("../data/processed/chicago_crime_clean.csv")

# Ensure datetime
df["date"] = pd.to_datetime(df["date"])

df.head()
```

```python
df["hour"] = df["date"].dt.hour
df["day_of_week"] = df["date"].dt.dayofweek  # Monday=0
df["month"] = df["date"].dt.month

df[["date", "hour", "day_of_week", "month"]].head()
```

```python
spatial_features = df[["latitude", "longitude"]]

spatio_temporal_features = df[["latitude", "longitude", "hour"]]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_spatial = scaler.fit_transform(spatial_features)

X_spatio_temporal = scaler.fit_transform(spatio_temporal_features)
```
